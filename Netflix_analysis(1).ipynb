# Top Actors on Netflix

# This step shows which actors appear the most in Netflix titles.
# We will count actor appearances by splitting the 'cast' column.


```python
# Drop missing values in 'cast' column
actor_series = df['cast'].dropna().str.split(', ')

# Flatten the list of all actors
from collections import Counter
actor_list = [actor for sublist in actor_series for actor in sublist]

# Count top 10 actors
top_actors = Counter(actor_list).most_common(10)

# Convert to DataFrame for plotting
import pandas as pd
top_actors_df = pd.DataFrame(top_actors, columns=['Actor', 'Count'])

print("Top 10 Actors on Netflix:\n", top_actors_df)

# Plot bar chart
plt.figure(figsize=(12,6))
plt.bar(top_actors_df['Actor'], top_actors_df['Count'], color='skyblue')
plt.title("Top 10 Actors Appearing on Netflix")
plt.xlabel("Actor")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45, ha='right')
plt.show()

```

# Duration Analysis (Movies Runtime & TV Show Seasons)

# This step analyzes how long Netflix content usually is:
# - For Movies  check typical runtime in minutes
# - For TV Shows  check number of seasons


```python
# Separate Movies and TV Shows
movies = df[df['type'] == 'Movie'].copy()
tv_shows = df[df['type'] == 'TV Show'].copy()

# --- Movies Runtime Analysis ---
# Clean up 'duration' column (remove 'min')
movies['duration'] = movies['duration'].str.replace(' min', '', regex=False)
movies['duration'] = pd.to_numeric(movies['duration'], errors='coerce')

# Plot histogram of movie runtimes
plt.figure(figsize=(10,5))
movies['duration'].dropna().hist(bins=30, color='orange')
plt.title("Distribution of Movie Runtimes on Netflix")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.grid(True)
plt.show()

# --- TV Shows Seasons Analysis ---
# Clean up 'duration' column (remove 'Season' or 'Seasons')
tv_shows['duration'] = tv_shows['duration'].str.replace(' Season', '', regex=False)
tv_shows['duration'] = tv_shows['duration'].str.replace(' Seasons', '', regex=False)
tv_shows['duration'] = pd.to_numeric(tv_shows['duration'], errors='coerce')

# Plot histogram of TV show seasons
plt.figure(figsize=(10,5))
tv_shows['duration'].dropna().hist(bins=15, color='purple')
plt.title("Distribution of TV Show Seasons on Netflix")
plt.xlabel("Number of Seasons")
plt.ylabel("Number of TV Shows")
plt.grid(True)
plt.show()

#  These charts help us understand:
# - How long most Netflix movies are (in minutes)
# - How many seasons most TV shows run for
```

```python
# STEP 15: Word Cloud of Titles 
# This step visualizes the most common words in Netflix titles.

from wordcloud import WordCloud

# Combine all titles into one string
title_words = " ".join(df['title'].dropna())

# Generate word cloud
wordcloud = WordCloud(width=1000, height=600, background_color="black", colormap="spring").generate(title_words)

# Plot word cloud
plt.figure(figsize=(12,8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Most Common Words in Netflix Titles", fontsize=16)
plt.show()

#  This word cloud highlights recurring themes/words in Netflix titles.
```
